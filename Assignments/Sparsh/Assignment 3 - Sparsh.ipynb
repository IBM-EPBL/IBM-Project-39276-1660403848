#### 6. Find the outliers and replace the outliers

```python
list_ = ['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings' ]
```

```python
df.quantile(q=(0.25,0.75))
```

```python
df.plot(kind="box",subplots=True, layout = (2,4), figsize=(20,15))
plt.show()
```

```python
import numpy as np
for x in list_:
    print("{:<15} - ".format(x),np.mean(df[x]))
```

```python
for x in list_:
    Q1 = df[x].quantile(0.25)
    Q3 = df[x].quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    df[x] = np.where(df[x] > upper, upper, df[x])
    df[x] = np.where(df[x] < lower, lower, df[x])
```

```python
df.plot(kind="box",subplots=True, layout = (2,4), figsize=(20,15))
plt.show()
```

#### 8. Split the data into dependent and independed variables

```python
y = df['Rings']
y.head(5)
```

```python
x = df.drop(columns=['Rings'], axis=1)
x.head(5)
```

#### 9. Scale the independent variables

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(x)
x = pd.DataFrame(X, columns = x.columns)
x
```

#### 10. Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=1)
```

#### 11. Build the model

```python
from sklearn.linear_model import LinearRegression
from sklearn import svm
LR = LinearRegression()
SVM = svm.SVR()
```

#### 12. Train the model

```python
LR.fit(x_train, y_train)
SVM.fit(x_train, y_train)
```

#### 13. Test the models

Linear Regression

```python
y_pred_LR = LR.predict(x_test)
y_pred_LR
```

SVM

```python
y_pred_SVM = SVM.predict(x_test)
y_pred_SVM
```

#### 14. Measure the performance using metrics

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

R2 Score

```python
r2_LR = r2_score(y_test,y_pred_LR)
print("R2 score of Linear Regression: ",r2_LR)
r2_SVM = r2_score(y_test, y_pred_SVM)
print("R2 score of SVM:               ",r2_SVM)
```

Mean Absolute Error

```python
mae_LR = mean_absolute_error(y_test,y_pred_LR)
print("Mean Absolute Error of Linear Regression: ",mae_LR)
mae_SVM = mean_absolute_error(y_test, y_pred_SVM)
print("Mean Absolute Error of SVM:",mae_SVM) 
```

Mean Squared Error

```python
mse_LR = mean_squared_error(y_test,y_pred_LR)
print("Mean Squared Error of Linear Regression: ",mse_LR)
mse_SVM = mean_squared_error(y_test, y_pred_SVM)
print("Mean Squared Error of SVM:",mse_SVM) 
```
