## 6. Find the outliers and replace the outliers

```python
df.quantile(q=(0.25,0.75))
```

```python
sns.boxplot(df['Age'])
```

```python
np.median(df['Age'])
```

```python
q1=df['Age'].quantile(0.25)
q3=df['Age'].quantile(0.75)
iqr=q3-q1
upper=q3+1.5*iqr
lower=q1-1.5*iqr
df['Age']=np.where(df['Age'] > upper, upper, df['Age'])
df['Age']=np.where(df['Age'] < lower, lower, df['Age'])
```

```python
sns.boxplot(df['Age'])
```
