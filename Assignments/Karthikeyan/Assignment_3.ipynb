### 6. Find the outliers and replace the outliers

```python
cols = ['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
```

```python
df.quantile(q = (0.25,0.75))
```

```python
df.plot(kind="box", subplots=True, layout=(7,4), figsize=(25,30))
```

```python
import numpy as np
print("Means of all columns\n")
for x in cols:
    print(x, ": ", np.mean(df[x]))
```

```python
for x in cols:
    q1 = df[x].quantile(0.25)
    q3 = df[x].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    df[x] = np.where(df[x] > upper, upper, df[x])
    df[x] = np.where(df[x] < lower, lower, df[x])
```

```python
df.plot(kind="box", subplots=True, layout=(7,4), figsize=(25,30));
```
